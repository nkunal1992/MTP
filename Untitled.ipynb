RGB Color image

Matplotlib ---> RGB

OpenCV ----> BGR

```python
cimg = cv2.imread('Narayan.jpg')
```

```python
cimg.shape
```

```python
plt.imshow(cimg)
```

We need to conver from BGR to RGB as Matplotlib expects RGB

```python
cimg1 = cv2.cvtColor(cimg,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cimg1)
```

Changing image dimension 

```python
w_ratio = 0.5 #20% of original width
h_ratio = 0.5 #20% of original height
```

```python
new_img1 = cv2.resize(cimg1,(0,0),cimg1,w_ratio,h_ratio)
```

```python
plt.imshow(new_img1)
```

```python
new_img1.shape
```

Flipping image

```python
new_img2 = cv2.flip(cimg1,0) # flip along horizontal axis
plt.imshow(new_img2)
```

```python
new_img3 = cv2.flip(cimg1,1) # flip along vertical axis
plt.imshow(new_img3)
```

```python
new_img4 = cv2.flip(cimg1,-1) # flip along both axis
plt.imshow(new_img4)
```

Saving image

```python
new_img4 = cv2.cvtColor(new_img4,cv2.COLOR_BGR2RGB)
cv2.imwrite('1Narayan.jpg',new_img4)
```
